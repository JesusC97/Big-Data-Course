````python
'''
  Autor: Alejandro Perdiguero
  Fecha: 10/04/2023
  Objetivo: Cálculo de Hipotenusa a partir de dos catetos.
'''

# PASO 1: Creamos variables
c1 = float(input("Dame el valor de cateto 1 "))
c2 = float(input("Dame el valor de cateto 2 "))
h = 0.0

# PASO 2: Ecuación
if (c1 > 0 and c2 > 0): # si los lados no son positivos, no hay triangulo
  h = (c1 ** 2 + c2 ** 2) ** 0.5
  print("La hipotenusa vale...", round(h,2), h)
else:
  print("No me valen esos datos :'( ")
````

## Ejercicio de Calculadora

### S para Salir: ('S')

### suma o s --> a + b
### resta o r --> a - b
### multiplica o m --> a*b
### divide o d --> a/b

````python
'''
  Autor: Jesus Cies
  Fecha: 16/10/2023
  Objetivo: Crear una calculadora.
'''
# Primero le pedimos al usuario que introduzca las variables de entrada (la operacion que quiere realizar)
# Solo le dispondremos de 4 operaciones: sumar, restar, multiplicacion y division.
# Incluimos como condición el caso del divisor=0 y, además, una palabra para poder salir.

operacion = input('Buenas tardes, ¿Qué operación quieres realizar?\nSi quieres que pare la calculadora tiene que introducir la letra S (mayus)\n[sumar: [s], restar: [r], multiplicacion: [m], d:[d]: ')

while (operacion != 'S'): # el bucle se inicia unicamente cuando no se ha puesto una S mayus

    a = float(input('Por favor, dame el valor de a: ')) # primer numero
    b = float(input('Gracias, ahora dame el valor de b: ')) # segundo numero

    if(operacion == 's'): # operacion suma
        print('el resultado es:', a + b)

    elif(operacion == 'r'): # operacion resta
        print('el resultado es:',a - b)

    elif(operacion == 'm'): # operacion multiplicacion
        print('el resultado es:',a * b)

    elif(operacion == 'd'): # operacion division
        if(b == 0):
            print('ERROR: DIV/0')
        else:
            print('el resultado es:',a/b)
    else:
      print('Lo siento, pero no sé que tipo de operación quieres realizar')
    operacion = input('¿Qué operación quieres realizar?\nSi quieres que pare la calculadora tiene que introducir la letra S (mayus)\n[sumar: [s], restar: [r], multiplicacion: [m], d:[d]: ')
````

### Respuestas


````python

# operacion = leerOperacion()
operacion = input('Dame la operación a realizar... ')

# MIENTRAS no sea S
while (not operacion =='S'): # operacion != 'S'
    # leemos A y B
    a = float(input('Dame el valor de A'))
    b = float(input('Dame el valor de B'))

    # si operacion = s
    if(operacion == 's' or operacion == 'suma'):
        # haz la suma
        print(a + b)
    # si operacion = r
    elif(operacion == 'r' or operacion == 'resta'):
        # haz la resta
        print(a - b)
    # si operacion = m
    elif(operacion == 'm' or operacion == 'multiplica'):
        # haz la multiplicación
        print(a * b)
    # si operación = d
    elif(operacion == 'd' or operacion == 'divide'):
        # haz la division (controlamos que no sea /0)
        if(b == 0):
            print('Error...')
        else:
            print(a / b)
    else:
      print('No sé de qué me hablas')
    operacion = input('Dame la operación a realizar... ')
````
