# Modificación de las series o columnas

Me doy cuenta visualizando los datos de mi tabla que en la serie o columna 'GoodPlayers' los datos son tipo texto (string) debido a que muestra los jugadores recomendados por la comunidad para jugar a un juego de mesa determinado.

No me interesa que sea un string, me interesa que sea una lista por mejorar su accesibilidad. Además, tengo interés en definir el tipo de variable que existe en cada serie por el mismo motivo.

```python
for i in range(len(board.GoodPlayers)): #Para cada valor de la serie 'GoodPlayers'
  board.GoodPlayers[i]=board.GoodPlayers[i].replace('[','').replace(']','').split(', ') # Le quito los caracteres de los corchetes para luego decir que es una lista separadas por (, )
display(board) # Muestro la tabla para comprobar visualmente que no he hecho nada raro (luego los compruebo mejor con otra funcion)
```

```python
# En este notebook realizo una comprobación de los elementos de la columna. Por si acaso ...

#board.GoodPlayers[0][0]=str(board.GoodPlayers[0][0]) # Funcion para convertir en string

board.GoodPlayers=board.GoodPlayers.str.replace('[','').str.replace(']','').str.split(', ') # Le quito los caracteres de los corchetes para luego decir que es una lista separadas por (, )

print(board.GoodPlayers[0][0])
print(board.GoodPlayers[0][1])
print(type(board.GoodPlayers)) # GoodPlayers es una columna
print(type(board.GoodPlayers[0])) # la columna tiene listas
print(type(board.GoodPlayers[0][0])) # la lista tiene strings (creo que es mejor dejarlo así debido a que hay muchos formatos de datos en esta columna (hay datos que tienen 4+ u otros que estan vacios))
```

Una vez modificado el formato de los datos de mis series según el formato que yo quiero decido ahora realizar una visualización de valores estadísticos descriptivos de las series.

Posteriormente realizo algunas gráficas para visualizar la frecuencia de algunos valores que me interesan.

```python
# El rango intercuartilico me habla de como están distribuidos los datos.
# Está definido como el tercer cuartil menos el primer cuartil.
# Calculo varios rangos para mostrarlos luego en gráficas.

Q1dificultad = board.GameWeight.quantile(0.25)
Q3dificultad = board.GameWeight.quantile(0.75)
print('El rango Intercuartilico de la dificultad de juegos de mesa es',Q3dificultad - Q1dificultad,'\n')

Q1rating = board.AvgRating.quantile(0.25)
Q3rating = board.AvgRating.quantile(0.75)
print('El rango Intercuartilico de la puntuacion media de juegos de mesa es',Q3rating - Q1rating,'\n')

Q1expansion = board.NumExpansions.quantile(0.25)
Q3expansion = board.NumExpansions.quantile(0.75)
print('El rango Intercuartilico del numero de expansiones de juegos de mesa es',Q3expansion - Q1expansion)
```

```python
IQRdificultad=Q3dificultad-Q1dificultad
maximoDificultad = Q3dificultad + 1.5*IQRdificultad
minimoDificultad = Q1dificultad - 1.5*IQRdificultad

print('El máximo de la distribución es', maximoDificultad,'\n')
print('El mínimo de la distribución es', minimoDificultad,'\n')

# NO me interesan los juegos que son excesivamente dificil o excesivamente fácil
# basicamente me dan tremenda pereza. Creo un filtro por prueba como estos parámetros.

filtroMaxDificultad = board.GameWeight > 4.5
filtroMinDificultad = board.GameWeight < 0.5

display(board[filtroMaxDificultad])
display(board[filtroMinDificultad])
```

```python
IQRrating=Q3rating-Q1rating
maximoRating = Q3rating + 1.5*IQRrating
minimoRating = Q1rating - 1.5*IQRrating

print('El máximo de la distribución es', maximoRating,'\n')
print('El mínimo de la distribución es', minimoRating,'\n')

# NO me fio de las valoraciones que son excesivamente bajas (haters) o excesivamente altas (fanboys)
# Creo un filtro por prueba como estos parámetros.

filtroMaxRating = board.AvgRating > maximoRating
filtroMinRating = board.AvgRating < minimoRating

display(board[filtroMaxRating])
display(board[filtroMinRating])
```

```python
IQRexpansion=Q3expansion-Q1expansion
maximoExpansion = Q3expansion + 1.5*IQRexpansion
minimoExpansion = Q1expansion - 1.5*IQRexpansion

print('El máximo de la distribución es', maximoExpansion,'\n')
print('El mínimo de la distribución es', minimoExpansion,'\n')

# Me gustan los juegos de mesa que tienen expansiones porque me gusta pillarlas todas.
# No quiero ver los juegos que no tienen expansiones o tienen demasiadas.
# Creo un filtro por prueba como estos parámetros.

filtroMaxExpansion = board.NumExpansions > 15
filtroMinExpansion = board.NumExpansions < 1

display(board[filtroMaxExpansion])
display(board[filtroMinExpansion])
```

## Realización de gráficas

En este apartado realizo algunas gráficas de varias series para visualizar mejor como están distribuidos los datos de dichas series.

```python
# Muestro con esta grafica la frencuencia de los juegos de mesa con una dificultad determinada
# en todo su rango de valores.

ax=board.GameWeight.plot.hist(bins=range(0, int(board.GameWeight.max()+1)))

# Le pongo titulos a los ejes horizontal y vertical además de un título del gráfico.
ax.set_xlabel('Dificultad de los juegos de mesa') # Titulo eje horizontal
ax.set_ylabel('Frecuencia') # Titulo eje vertical
ax.set_title('Histograma de dificultad de los juegos de mesa') # Titulo del grafico
```

```python
# Muestro ahora la puntuación media en un histograma.

ax=board.AvgRating.plot.hist(bins=range(0, int(board.AvgRating.max()+1)))

# Le pongo titulos a los ejes horizontal y vertical además de un título del gráfico.

ax.set_xlabel('Puntuación de los juegos de mesa')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma de puntuación media de los juegos de mesa')
```

```python
# Me parece muy raro que no se vea nada (aparentemente) entre el 1.0 y 2.0.
# Quiero comprobarlo poniendo un filtro.

filtroRating1 = board.AvgRating > 1.0 and board.AvgRating < 2.0  # ESTO NO FUNCIONA
filtroRating2 = board.AvgRating > 2.0 and board.AvgRating < 3.0

display(board[filtroRating1])
display(board[filtroRating2])
```
